### 2. 3D 曲面畫圖

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='coolwarm',
                linewidth=0,  antialiased=False)
```

### 3. Level maps and Contour maps

```python
x = np.arange(-5, 5, 0.01)
y = np.arange(-5, 5, 0.01)
```

```python
X, Y = np.meshgrid(x, y)
Z = -3*Y / (X**2+Y**2+1)
```

```python
plt.contour(X, Y, Z)
```

應用: 畫隱函數

```python
x = np.linspace(-3.5, 2, 200)
y = np.linspace(-2.5, 2.5, 200)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = Y**2 - X**3 - 3*X**2
```

```python
plt.contour(X, Y, Z, [0])
```
