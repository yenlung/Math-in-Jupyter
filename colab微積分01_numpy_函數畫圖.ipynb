## 2. 函數畫圖基本篇

基本上就是描點法。

```python
x = np.linspace(-5, 5, 200)
y = x**2
```

```python
plt.plot(x,y)
```

可以自己多試一些函數。

```python
x = np.linspace(-10, 10, 200)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

## 3. 進階挑戰

如果畫個不連續的函數...

```python
x = np.linspace(-2, 2, 1000)
y = 1/x
```

```python
y[np.abs(y)>10] = np.nan
```

```python
plt.plot(x, y)
```

那 $y = \tan(x)$ 可以嗎?

```python
x = np.linspace(-6.28, 6.28, 1000)
y = np.tan(x)
```

```python
y[np.abs(y)>10]=np.nan
```

```python
plt.plot(x,y)
```

## 4. 用 `scatter` 畫點

```python
xx = np.linspace(-5, 5, 100)
yy = xx**2
```

```python
plt.plot(xx, yy)
plt.scatter([0], [0], c='r', zorder=3)
```

```python
plt.plot(xx, yy, c='k')
plt.scatter([0], [0], c='w', s=50, edgecolors='k', zorder=3)
plt.scatter([0], [5], c='k', s=50)
```
