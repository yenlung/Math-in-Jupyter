## 2. 有 $y \to \infty$ 的點該如何

如果我們要畫的圖, 有當 $x \to a^-$, 或 $x \to a^+$ 時, y 會趨近 $+\infty$ 或 $-\infty$ 的情況, Python 會畫出什麼呢?

我們試著畫

$$y = \tan(x)$$

試試。因為

$$\tan(x) = \dfrac{\sin(x)}{\cos(x)}$$

所以當 $\cos(x) = 0$ 時, $\tan(x)$ 是沒有定義的。換言之, 當

$$x = \cdots, -\frac{3\pi}{2}, -\frac{\pi}{2}, \frac{\pi}{2}, \frac{3\pi}{2},\cdots$$

時, $\tan(x)$ 沒有定義。這種情況我們要 Python 畫圖會如何呢?

```python
y = np.tan(x)
```

```python
plt.plot(x, y)
```

我們發現 Python 會畫出不知是什麼東西的圖, 這是因為那些很大的數字, 會拉到很遠去。更嚴重的是, 本來不連續的點還會連起來!! 

解決的方案是, 我們可以告訴 Python, 絕對值大於某個數的就不要了! 比方說, 絕對值大於 20 的數字不要是這麼寫。

這裡 `np.nan` 是說這個值不是一個數字, 於是在描點時會當沒有這個點。

```python
y[np.abs(y)>20] = np.nan
```

```python
plt.plot(x, y)
```

就這樣, 我們就可以畫出正確的 $\tan(x)$ 函數圖形。用這個技巧, 你可以試著去畫其他有不連續函數的圖, 例如  $f(x) = 1/x$ 等等。
