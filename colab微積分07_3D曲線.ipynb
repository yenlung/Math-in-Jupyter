這次我們來畫 $\mathbb{R}^3$ 裡面的曲線! 

```python
%matplotlib inline

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

## 1. 

一般在三維空簡的曲線, 我們是以參數式表示。比如 $x=\cos t, y= \sin t, z=t$

```python
t = np.linspace(0, 5*np.pi, 500)

x = np.cos(t)
y = np.sin(t)
z = t

plt.axes(projection='3d')
plt.plot(x, y, z)
```

我們發現只要設為 `projection='3d'` 就可以了!

注意, 舊版的 `matplotlib` 需要先

`from mpl_toolkits import mplot3d`

才可以, 新版 (3.2.0 之後) 就不用了。

## 2.

順道一提, 和一般 `matplotlib` 二維一樣, 就 `plot`, `scatter` 兩大系統

```python
t = np.linspace(0, 5*np.pi, 500)

x = np.cos(t)
y = np.sin(t)
z = t

u = np.random.rand(100) * 5*np.pi
xp = np.cos(u) + 0.1*np.random.randn(100)
yp = np.sin(u) + 0.1*np.random.randn(100)
zp = u + 0.1*np.random.randn(100)

ax = plt.axes(projection='3d')
plt.plot(x, y, z)
ax.scatter(xp, yp, zp, c=np.random.randint(0, 5, 100), s=30)
```
